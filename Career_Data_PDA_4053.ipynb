To remove the outliners using univariant imputation by usin statical method Mean, Mode, Median. By using seaborn we can identify the outliner for Age,
Years of Experience,Job Opportunities and salary.

```python
#Convert selected columns to numeric properly
cols_to_convert = ['Age', 'Years of Experience', 'Job Opportunities', 'Salary']
for col in cols_to_convert:
    career_data[col] = pd.to_numeric(career_data[col], errors='coerce')
```

```python
# To eremove NAN
career_data = career_data.dropna()
```

```python
# After removing check whether the dataset is null
career_data.isnull().sum()
```

We will use mean() method as there is no outliners

```python
plt.figure(figsize= (30,6))
plt.subplot(1, 4, 1)
sns.boxplot(career_data["Age"])
plt.subplot(1, 4, 2)
sns.boxplot(career_data["Years of Experience"])
plt.subplot(1, 4, 3)
sns.boxplot(career_data["Job Opportunities"])
plt.subplot(1, 4, 4)
sns.boxplot(career_data["Salary"])
plt.show()
```

Refer to above box plot outliners are seen on Salary column we are using median() method.
For Age, Years of Experince, and Job oppurtunities use mean() method

```python
# Mean() for Age
age_value = career_data["Age"].mean()
age_value
```

```python
# Mean() for Years of Experince, and Job oppurtunities
yrs_experince = career_data["Years of Experience"].mean()
yrs_experince
```

```python
# Mean() for job oppurtunities
job_oppurtunities = career_data["Job Opportunities"].mean()
job_oppurtunities
```

```python
# To use fillna method for Age, Years of Experience and Job oppurtunities
career_data.loc[:, "Age"] = career_data["Age"].fillna(age_value)
career_data.loc[:, "Years of Experience"] = career_data["Years of Experience"].fillna(yrs_experince)
career_data.loc[:, "Job Opportunities"] = career_data["Job Opportunities"].fillna(job_oppurtunities)
```

```python
career_data
```

```python
# In the process of fixing outliner with median()
median_salary = career_data['Salary'].median()
```

```python
# To Use fillna method to fill the Salary
career_data.loc[:, "Salary"] = career_data["Salary"].fillna(median_salary)
```

```python
# Using IQR method to identify the upper and lower limit in Salary
Q1 = career_data['Salary'].quantile(0.25) # Any value below 0.25 is outliners
Q3 = career_data['Salary'].quantile(0.75) # Any value above 0.75 is outliners
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

#career_data = career_data[(career_data['Salary'] >= lower_limit) & (career_data['Salary'] <= upper_limit)]
```

```python
# To fix outliner use clip method
career_data.loc[:, "Salary"] = career_data["Salary"].clip(lower_limit, upper_limit)
```

Now to use encoding method for categogerical data, to convert to numerical data; before encoding use mode() method to fix the NAN in the data set columns of
Field of Study, Current Occupation, Gender and	Education Level.

```python
# To convert Gender value in to numeric
career_data = pd.get_dummies(career_data, columns=["Gender"], dtype="int64", drop_first=True)
```

```python
# Check if 'Gender' column exists before encoding
if 'Gender' in career_data.columns:
    # To convert Gender value in to numeric
    career_data = pd.get_dummies(career_data, columns=["Gender"], dtype="int64", drop_first=True)
    print("One-hot encoding for 'Gender' applied.")
else:
    print("Error: 'Gender' column not found in the DataFrame.")
```
